```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

purple = np.array(pd.read_excel('Purple.xlsx', usecols='AN:AP'))
data = np.array(pd.read_excel('compressed.xlsx',usecols='BE:BI')) #slime green

x, y = data[:,0], data[:,4] #extracting raw x and Y values from 1st and second column of sheet

LOWER_BOUND = 400  #smallest wavelength value to interpolate
UPPER_BOUND = 800  #largest wavelength value to interpolate
numOfValues = UPPER_BOUND - LOWER_BOUND + 1 #number of values to produce within the bounds


x_space = np.linspace(LOWER_BOUND,UPPER_BOUND, numOfValues ) #whole number array list w/ +1 increments
y_space = np.interp(x_space, x, y) #interpolating using whole number values and x, y data from spreedsheet


plt.plot(x_space, y_space)
plt.title('Interpolated slime green')
plt.xlabel('Wave Length')
plt.ylabel('% Reflectance')
plt.show()
```
